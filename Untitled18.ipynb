```python
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df.head()
```

```python
display(df.head())
display(df.isnull().sum())
display(df.info())
```

```python
median_income = df['median_income']
median_house_value = df['median_house_value']
```

```python
import matplotlib.pyplot as plt

plt.scatter(median_income, median_house_value)
plt.title("Relationship between Median Income and Median House Value")
plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.show()
```

```python
housing_age_counts = df['housing_median_age'].value_counts()
display(housing_age_counts)
```

```python
housing_age_bins = [0, 10, 20, 30, 40, 50, df['housing_median_age'].max()]
housing_age_labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51+']
df['housing_age_group'] = pd.cut(df['housing_median_age'], bins=housing_age_bins, labels=housing_age_labels, right=True, include_lowest=True)
housing_age_group_counts = df['housing_age_group'].value_counts()
display(housing_age_group_counts)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(housing_age_group_counts, labels=housing_age_group_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Housing Median Age Groups')
plt.axis('equal')
plt.show()
```

```python
# Drop the non-numeric 'housing_age_group' column before calculating the correlation matrix
correlation_matrix = df.drop('housing_age_group', axis=1).corr()
display(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Housing Data')
plt.show()
```
